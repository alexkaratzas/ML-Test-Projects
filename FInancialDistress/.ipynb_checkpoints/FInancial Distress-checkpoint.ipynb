```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix
```

```python
def print_scores(y_test_, y_pred_):
    print(f'Accuracy: {accuracy_score(y_test_, y_pred_)}')
    print(f'Precision: {precision_score(y_test_, y_pred_)}')
    print(f'Recall: {recall_score(y_test_, y_pred_)}')
    print(f'ROC AUC: {roc_auc_score(y_test_, y_pred_)}')
    tn, fp, fn, tp = confusion_matrix(y_test_, y_pred_).ravel()
    print('TP: {0: >5d}   FP {1: >5d}'.format(tp, fp))
    print('TN: {0: >5d}   FN {1: >5d}'.format(tn, fn))

print_scores(y_test, y_pred)
```

# Data Pre-processing

### One-hot encode categorical feature

```python
df.shape
```

```python
df = pd.get_dummies(df, prefix=['x80_'],columns=['x80'])
```

```python
df.shape
```

### Normalize and regularize numeric features

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.columns[:-38]
```

```python
stdScaler = StandardScaler()
columns_to_scale = df.columns[:-38]
df[columns_to_scale] = stdScaler.fit_transform(df[columns_to_scale])
```

```python
import math
from matplotlib.colors import ListedColormap
cm3 = ListedColormap(['#0000aa', '#ff2020', '#50ff50'])
cm3(0)

fig, axes = plt.subplots(math.ceil(len(origDF.columns)/4), 4, figsize=(30,30))
defaulted = origDF[origDF.Label == 0]
notdefaulted = origDF[origDF.Label == 1]
ax = axes.ravel()

for i in range(len(origDF.columns)):
    _, bins = np.histogram(origDF.iloc[:, i],bins=50)
    ax[i].hist(defaulted.iloc[:,i], bins=bins,color=cm3(0), alpha= .5)
    ax[i].hist(notdefaulted.iloc[:,i], bins=bins, color=cm3(2), alpha=.5)
    ax[i].set_title(origDF.columns[i])
    ax[i].set_yticks(())
ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["Defaulted", "Not Defaulted"], loc="best")
fig.tight_layout()
 
```
