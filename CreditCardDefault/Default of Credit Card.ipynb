# Histograms colored by class label

```python

from matplotlib.colors import ListedColormap
cm3 = ListedColormap(['#0000aa', '#ff2020', '#50ff50'])
cm3(0)
```

```python
import math

fig, axes = plt.subplots(math.ceil(len(df.columns)/2), 2, figsize=(20,30))
defaulted = df[df.LABEL == 0]
notdefaulted = df[df.LABEL == 1]
ax = axes.ravel()

for i in range(len(df.columns)):
    _, bins = np.histogram(df.iloc[:, i],bins=50)
    ax[i].hist(defaulted.iloc[:,i], bins=bins,color=cm3(0), alpha= .5)
    ax[i].hist(notdefaulted.iloc[:,i], bins=bins, color=cm3(2), alpha=.5)
    ax[i].set_title(col_names[i])
    ax[i].set_yticks(())
ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["Defaulted", "Not Defaulted"], loc="best")
fig.tight_layout()
 
```

# Models

```python
from sklearn.model_selection import train_test_split 
```

```python
df = df.drop(['ID'], axis=1)
df
```

```python
X = df.values
y = df['LABEL'].values
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train.shape
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
len(y_pred[y_pred == 1])
```

```python
len(y_test[y_test == 1])
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
roc_auc_score(y_test, y_pred)
```
